```python
# 각 자리 수의 합을 구하는 EachSum 클래스의 compute() 메소드 작성
class EachSum:
    def __init__(self):
        self.total = 0  
    def compute(self, num):
        for i in str(num):
            self.total += int(i)
        return self.total
# 코드 작성

each = EachSum()
total = each.compute(123456)
print('합계', total)
```

```python
class Calculator:
    # 생성자 작성
    def __init__(self, first, second):
        self.first = first
        self.second = second
    def plus(self):
        return self.first + self.second
    def minus(self):
        return self.first - self.second
    def multiply(self):
        return self.first * self.second
    def divide(self):
        return self.first / self.second

cal = Calculator(1, 5)
cal.plus()
```

```python
class Pizza:
    
    # 생성자 작성
    def __init__(self):
        self.topping = []
    # add_topping 메소드 작성
    def add_topping(self, topping):
        self.topping += [topping]
    # remove 메소드 작성
    def remove(self, topping):
        self.topping.remove(topping)
    # show_topping 메소드 작성
    def show_topping(self):
        return self.topping

p1 = Pizza()
p1.add_topping('치즈')
p1.add_topping('베이컨')
p1.add_topping('고구마')
p2 = Pizza()
p2.add_topping('소시지')
p2.add_topping('불고기')
p2.add_topping('피망')
p2.remove('소시지')
print(p1.show_topping())
print(p2.show_topping())
```

```python
#연습문제 파이썬 11_01 create
import pymysql

conn = None

try:
    conn = pymysql.connect(
    host='localhost', user='root', password='3318', db='python', charset='utf8')
    
    with conn.cursor() as cursor:
        sql = '''CREATE TABLE article(id INTEGER PRIMARY KEY AUTO_INCREMENT, title VARCHAR(30), content VARCHAR(30), writer VARCHAR(30), hit INT, c_date DATETIME)'''
        cursor.execute(sql)
        conn.commit()
       
    
except pymysql.InternalError as e :
    print('예외 발생',e)
finally:
    if conn: #conn의 값이 있다면~
        conn.close() #try 내에만 있어도 전역변수처럼 쓸 수 있지만, try 밖에서 정확히 conn을 선언하여 전역변수로 명시해주는게 좋음
```

```python
#SELECT fetchone
import pymysql

conn = None

try:
    conn = pymysql.connect(
    host='localhost', user='root', password='3318', db='python', charset='utf8')
    
    cursor = conn.cursor()
    sql = '''SELECT * FROM mysql'''
    cursor.execute(sql)
    result = cursor.fetchone() #한줄만 나옴
    print(result) #튜플 한개
    print(result[0]) #튜플의 첫번째 요소
    
    conn.commit()
    cursor.close()
    conn.close()
    
except pymysql.InternalError as e :
    print('예외 발생',e)
```

```python
#SELECT fetchmany
import pymysql

conn = None

try:
    conn = pymysql.connect(
    host='localhost', user='root', password='3318', db='python', charset='utf8')
    
    cursor = conn.cursor()
    sql = '''SELECT * FROM mysql'''
    cursor.execute(sql)
    result = cursor.fetchmany(2) #두줄까지만 나옴
    print(result) #튜플안에 튜플형태
    print(result[0]) #튜플형태
    print(result[0][1]) #튜플의 두번째 요소
    
    conn.commit()
    cursor.close()
    conn.close()
    
except pymysql.InternalError as e :
    print('예외 발생',e)
```

```python
#SELECT fetchall
import pymysql

conn = None

try:
    conn = pymysql.connect(
    host='localhost', user='root', password='3318', db='python', charset='utf8')
    
    cursor = conn.cursor()
    sql = '''SELECT * FROM mysql'''
    cursor.execute(sql)
    result = cursor.fetchall()
    print(result) #튜플안에 튜플형태
    print(result[0]) #튜플형태
    print(result[0][1]) #튜플의 두번째 요소
    
    conn.commit()
    cursor.close()
    conn.close()
    
except pymysql.InternalError as e :
    print('예외 발생',e)
```

```python
#delete
import pymysql

conn = None

try:
    conn = pymysql.connect(
    host='localhost', user='root', password='3318', db='python', charset='utf8')
    
    cursor = conn.cursor()
    sql = '''DELETE FROM mysql
             WHERE ID = %s'''
    cursor.execute(sql, (2)) #ID 2번 삭제
    conn.commit()
    cursor.close()
    conn.close()
    
except pymysql.InternalError as e :
    print('예외 발생',e)
```

```python
#update
import pymysql

conn = None

try:
    conn = pymysql.connect(
    host='localhost', user='root', password='3318', db='python', charset='utf8')
    
    cursor = conn.cursor()
    sql = '''UPDATE mysql SET TITLE = %s
             WHERE ID = %s'''
    cursor.execute(sql, ('제목수정됨', 1)) #ID 1번
    conn.commit()
    cursor.close()
    conn.close()
    
except pymysql.InternalError as e :
    print('예외 발생',e)
```
